```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set()
tips = sns.load_dataset('tips')
```

```python
# Distribution of tip amount 
sns.distplot(tips['tip'], kde=False)
```

```python
# Distribution of total amount
sns.set_style('whitegrid')
sns.distplot(tips['total_bill'], color='red', bins = 10,kde=False)
```

### Categorical plots

```python
sns.barplot(data=tips, x='day', y = 'total_bill', hue='sex', estimator=np.sum, ci = False)
```

```python
sns.barplot(data=tips, x='sex', y = 'total_bill', hue='day', estimator=np.sum, ci = False)
```

```python
res = sns.barplot( x = 'day', data=tips, y ='total_bill', hue='smoker', estimator=np.sum, ci = False)
```

```python
res = sns.catplot(data=tips, x = 'day', y = 'total_bill',col='sex',kind='bar',ci=False)
```

```python
res = sns.catplot(data=tips, x = 'sex', y = 'total_bill',col='day',kind='bar', hue='smoker',
                  col_wrap=2,ci=False)
```

```python
res = sns.boxplot(x = 'day',y ='total_bill', data=tips, showfliers=False)
plt.gcf().set_size_inches(15,8)
```

```python
res = sns.catplot(data=tips, x = 'day', y = 'total_bill', hue='smoker', col='sex', kind='box', showfliers=False)
```

```python
sns.stripplot(x='day',y='size', data = tips)
plt.gcf().set_size_inches(10,8)
```

```python
sns.set(style='whitegrid')
sns.swarmplot(x='day',y='size', data = tips)
plt.gcf().set_size_inches(20,8)
```

```python
sns.set(style='white')
sns.swarmplot(x='day',y='size', hue='smoker', data = tips)
plt.gcf().set_size_inches(20,8)
```

```python
sns.violinplot(x='day',y='total_bill', data = tips, inner=None)
plt.gcf().set_size_inches(20,8)
```

```python
sns.violinplot(x='day',y='total_bill', data = tips, inner=None)
sns.swarmplot(x='day',y='total_bill', color='white', data = tips)
plt.gcf().set_size_inches(20,10)
```

```python
sns.set_style('ticks')
sns.countplot(x='day', data = tips)
```

### Relationship plots

```python
sns.set_style('dark')
sns.scatterplot(data=tips, x = 'total_bill', y = 'tip')
```

```python
sns.set_style('dark')
sns.scatterplot(data=tips, x = 'total_bill', y = 'tip', hue="sex")
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', col='day', row='time')
```

### Regression Plot

```python
res = sns.regplot(data=tips, x = 'total_bill', y = 'tip')
```

```python
res = sns.lmplot(data=tips, x = 'total_bill', y = 'tip', col='day')
plt.gcf().set_size_inches(20,10)
```

## Pair plots  - Bivariate Relationships

```python
sns.pairplot(tips)
plt.gcf().set_size_inches(20,10)   # Set figure size 
```

```python
sns.pairplot(tips, hue='sex', markers=['+','o'])
plt.gcf().set_size_inches(20,10)   # Set figure size 
```
